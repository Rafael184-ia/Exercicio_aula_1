#Pedir numero e mostrar tabuada

```python
numero = int(input("digite um numero para ver a tabuada: "))
for tabuada in range(1 , 11):
    print(numero," +" , tabuada, "=",numero + tabuada)
```

#idade e documento

```python
idade_da_pessoa = int(input("Digite sua idade : "))
tem_documento = input("Tem documento ? (sim/nao): ")
if idade_da_pessoa >= 18 and tem_documento == "sim":
    print ("Entrada permitida")
else :
    print ("Entrada recusada")
```
